```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
txt_file = '/Users/yufanwang/Desktop/Study/Project/Data/Fscore.txt'
```

```python
name = []
weight = []
mean = []


TP = []
FP = []
FN = []
Prec = []
Rec = []
F05 = []
f = open(txt_file)               # 返回一个文件对象 
lines = f.readlines()               # 调用文件的 readline()方法 
name_lines = lines[0::7]
score_lines = lines[4::7]

for item in name_lines:
    value = item.split('_')
    
    name.append(value[0]+"_"+value[1])
    mean.append(value[2])
    weight.append(value[3])
exp=range(len(name_lines))

for item in score_lines:
    value = item.split('\t')
    TP.append(value[0])
    FP.append(value[1])
    FN.append(value[2])
    Prec.append(value[3])
    Rec.append(value[4])
    F05.append(value[5][:-2])
```

```python
data = {'name':name,
        'epoch':exp,
        'mean':mean,
    'weight':weight,
       'TP':TP,
       'FP':FP,
        'FN':FN,
        'Prec':Prec,
        'Rec':Rec,
        'F0.5':F05}
df = pd.DataFrame(data)
df
```
