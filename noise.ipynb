```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
old_file = '/Users/yufanwang/Desktop/Study/Project/NoiseGram/results/count_noise.txt'
new_file = '/Users/yufanwang/Desktop/Study/Project/Data/count_gau_1.5_noise.txt'
nn_file = '/Users/yufanwang/Desktop/Study/Project/Data/count_gau_0.1_noise.txt'
```

```python
weight = []
avg_count = []
avg_edit_rate = []

f = open(old_file)               # 返回一个文件对象 
lines = f.readlines()               # 调用文件的 readline()方法 
for item in lines[2::3]:
    value = item.split('/')
#     print(value[11])
    weight.append(float(value[11][19:]))
    sub_value = value[12].split(' ')
#     print(sub_value)
    avg_count.append(int(sub_value[1])/2755)
    avg_edit_rate.append(float(sub_value[2][:-2])/2755)
# #     align.append(value[4])

data = {'weight':weight,
       'avg edits':avg_count,
       'avg edit rate':avg_edit_rate}
df_old = pd.DataFrame(data)
df_old
```

```python
weight = []
avg_count = []
avg_edit_rate = []

f = open(new_file)               # 返回一个文件对象 
lines = f.readlines()               # 调用文件的 readline()方法 
for item in lines[2::3]:
    value = item.split('/')
#     print(value)
    print(value[12][-3:])
    weight.append(float(value[12][-3:]))
    sub_value = value[13].split(' ')
#     print(sub_value)
    avg_count.append(int(sub_value[1])/2755)
    avg_edit_rate.append(float(sub_value[2][:-2])/2755)
# # #     align.append(value[4])
```

```python
data = {'weight':weight,
       'avg edits':avg_count,
       'avg edit rate':avg_edit_rate}
df_new = pd.DataFrame(data)
df_new
```

```python
weight = []
avg_count = []
avg_edit_rate = []

f = open(nn_file)               # 返回一个文件对象 
lines = f.readlines()               # 调用文件的 readline()方法 
for item in lines[2::3]:
    value = item.split('/')
#     print(value[11])
    weight.append(float(value[11][19:]))
    sub_value = value[12].split(' ')
#     print(sub_value)
    avg_count.append(int(sub_value[1])/2755)
    avg_edit_rate.append(float(sub_value[2][:-2])/2755)
# #     align.append(value[4])

data = {'weight':weight,
       'avg edits':avg_count,
       'avg edit rate':avg_edit_rate}
df_nn = pd.DataFrame(data)
df_nn
```

```python
import matplotlib.pyplot as plt
# from matplotlib.pyplot import MultipleLocator

# x_major_locator=MultipleLocator(5)

fig, ax = plt.subplots(dpi=600,figsize=(6,4))

ax.plot(df_old['weight'], df_old['avg edits'], marker='o',markersize=2,label='Gramformer') # 作y1 = x 图，并标记此线名为linear
ax.plot(df_new['weight'], df_new['avg edits'], marker='o',markersize=2,label='GramNoise-Gau (1.5)') # 作y1 = x 图，并标记此线名为linear
ax.plot(df_nn['weight'], df_nn['avg edits'], marker='o',markersize=2,label='GramNoise-Gau (0.1)') # 作y1 = x 图，并标记此线名为linear

ax.set_xlabel('Sigma Value') #设置x轴名称 x label
ax.set_ylabel('Average Edits') #设置y轴名称 y label
# ax.set_title('The Effect of Adding Gaussian Noise on Gramformer Model') #设置图名为Simple Plot
ax.legend() #自动检测要在图例中显示的元素，并且显示
# plt.xlim((0, 15))
# ax.xaxis.set_major_locator(x_major_locator)
plt.show() #图形可视化
```
