```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
filePath = '/Users/yufanwang/Desktop/Study/Project/models/tensorboard'
files = sorted(os.listdir(filePath))
files
```

```python
out_former = pd.read_csv(filePath+'/'+files[-1])
out_former[['Step',"Value"]].head()
```

```python
adv = pd.read_csv(filePath+'/'+files[1])
adv[['Step',"Value"]].head()
```

```python
mean = pd.read_csv(filePath+'/'+files[3])
mean[['Step',"Value"]].head()
```

```python
fig, ax = plt.subplots(dpi=600,figsize=(9,6))
ax.plot(out_former['Step'], out_former['Value'].round(2),markersize=2,label='Gramformer') # y1 = x linear
ax.plot(adv['Step'], adv['Value'].round(2),markersize=2,label='GramAdv') # y1 = x linear
ax.plot(mean['Step'], mean['Value'].round(2), markersize=2,label='GramAdvMean') # y1 = x linear

ax.set_xlabel('Step') #x x label
ax.set_ylabel('Loss') #y y label
ax.legend() #
plt.show() #
```
