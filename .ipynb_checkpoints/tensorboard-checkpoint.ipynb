```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
filePath = '/Users/yufanwang/Desktop/Study/Project/models/tensorboard'
files = sorted(os.listdir(filePath))
files
```

```python
out_former = pd.read_csv(filePath+'/'+files[-1])
out_former[['Step',"Value"]].head()
```

```python
adv = pd.read_csv(filePath+'/'+files[1])
adv[['Step',"Value"]].head()
```

```python
mean = pd.read_csv(filePath+'/'+files[3])
mean[['Step',"Value"]].head()
```

```python
fig, ax = plt.subplots(dpi=600,figsize=(9,6))
ax.plot(out_former['Step'], out_former['Value'].round(2),markersize=2,label='Gramformer') # 作y1 = x 图，并标记此线名为linear
ax.plot(adv['Step'], adv['Value'].round(2),markersize=2,label='GramAdv') # 作y1 = x 图，并标记此线名为linear
ax.plot(mean['Step'], mean['Value'].round(2), markersize=2,label='GramAdvMean') # 作y1 = x 图，并标记此线名为linear

ax.set_xlabel('Step') #设置x轴名称 x label
ax.set_ylabel('Loss') #设置y轴名称 y label
ax.legend() #自动检测要在图例中显示的元素，并且显示
plt.show() #图形可视化
```
